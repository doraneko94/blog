```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
def f(x):
    return 1 / np.sqrt(x*2+np.square(x))
```

```python
xt = np.linspace(0, 1, 100)
yt = f(xt[1:])
```

```python
fig, ax = plt.subplots(1, 3, figsize=(12, 3))
for i, b in enumerate([10, 30, 60]):
    x = np.linspace(0, 1, b+1)
    y = f(x[1:])
    ax[i].plot(xt[1:], yt, label="f(x)")
    ax[i].bar((x[1:] + x[:-1]) / 2, y, x[1] - x[0], color="C1", edgecolor="k", alpha=0.5, label=r"$1/n\times f(k/n)$")
    ax[i].set_title(r"$n={}$".format(b))
    ax[i].set_xlabel("x")
    ax[i].set_ylabel("y")
    ax[i].legend()
plt.tight_layout()
plt.savefig("pwq.jpg")
```
